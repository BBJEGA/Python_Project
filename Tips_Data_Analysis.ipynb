```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

What's the average tip amount?

```python
df['tip'].mean()
```

Do men or women tip more?

```python
df.groupby('sex')['tip'].mean()
```

Which day has the highest total bill?

```python
df.groupby('day')['total_bill'].sum()
```

Create a Bar Plot for Average Tips by Gender

```python
sns.barplot(x='sex',y='tip',data=df)
plt.title('Average Tips by Gender')
plt.show()
```

Create Scatter Plot for Total Bill vs. Tip

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('Total Bill vs Tip')
plt.show()
```
